```python
import osmnx as ox, geopandas as gpd
%matplotlib inline
ox.config(log_file=True, log_console=True, use_cache=True)

# define a bounding box in Bogot
# north, south, east, west = 4.600276, 4.595870, -74.072767, -74.080487
north, south, east, west = 4.602276, 4.593870, -74.070767, -74.083487

# create network from that bounding box
G1 = ox.graph_from_bbox(north, south, east, west, network_type='drive_service')
G1 = ox.project_graph(G1)
fig, ax = ox.plot_graph(G1, node_size=30, save=True, filename='PlazaBolivarExampleGraph', file_format='svg')
```
